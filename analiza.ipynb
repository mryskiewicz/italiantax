```python
with open("kodeks_clean.txt", "r", encoding="utf-8") as f:
    text_cleaned = f.read()
```

```python
print(text_cleaned)
```

```python
import spacy

nlp = spacy.load("it_core_news_sm")
```

```python
import json

output_file = "tagged_articles_spacy.json"

articles = text_cleaned.split("\n\n")  # Przyjmujemy, że artykuły są oddzielone pustą linią

tagged_articles = []
for article in articles:
    doc = nlp(article)
    tagged_sentences = []

    for sent in doc.sents:
        tagged_words = [
            {
                "text": token.text,
                "lemma": token.lemma_,
                "pos": token.pos_,
                "tag": token.tag_
            }
            for token in sent
        ]
        tagged_sentences.append(tagged_words)

    tagged_articles.append({"article": article[:50], "content": tagged_sentences})  # Zapisujemy fragment artykułu dla kontekstu

with open(output_file, "w", encoding="utf-8") as f:
    json.dump(tagged_articles, f, ensure_ascii=False, indent=2)

print(f"Tagowanie zakończone. Wynik zapisano w pliku {output_file}")
```

```python
nouns = 0
verbs = 0
```

```python
for article in tagged_articles:
    for sentence in article["content"]:
        for word in sentence:
            if word["pos"] == "NOUN":
                nouns += 1
            elif word["pos"] == "VERB":
                verbs += 1
```

```python
nom_index = nouns / verbs
```

```python
print(f"nouns: {nouns}, verbs: {verbs}, index: {round(nom_index, 2)}")
```

```python
wider_nouns = 0
wider_verbs = 0
```

```python
for article in tagged_articles:
    for sentence in article["content"]:
        for word in sentence:
            if word["pos"] in ["NOUN", "PROPN", "PRON", "ADJ", "NUM"]:
                wider_nouns += 1
            elif word["pos"] in ["VERB", "AUX"]:
                wider_verbs += 1
```

```python
print(f"wider nouns: {wider_nouns}, wider verbs: {wider_verbs}, index: {round(wider_nouns / wider_verbs, 2)}")
```
